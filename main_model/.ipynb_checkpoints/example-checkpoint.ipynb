```python
import numpy as np
import pandas as pd# Update default settings to show 2 decimal place

column_names = ["x","y","z","weight", "l", "w","h","rotation"]
df = pd.DataFrame(columns = column_names)
```

```python
df
```

```python
for item in packer.item_strings:
    item_charac = item.split(",")
    print(item_charac)
    df2 = {'l': item_charac[0], 'w': item_charac[1], 'h': item_charac[2], 'weight': item_charac[3],'x': item_charac[4],'y': item_charac[5],'z': item_charac[6],'rotation': item_charac[7]}
    df = df.append(df2, ignore_index = True) 
```

```python
df
```

```python
df["x"] = df["x"].str.replace("Decimal","")
df["x"] = df["x"].str.replace("[","")
df["x"] = df["x"].str.replace("(","")
df["x"] = df["x"].str.replace(")","")
df["x"] = df["x"].str.replace("'","")

df["y"] = df["y"].str.replace("Decimal","")
df["y"] = df["y"].str.replace("[","")
df["y"] = df["y"].str.replace("(","")
df["y"] = df["y"].str.replace(")","")
df["y"] = df["y"].str.replace("'","")

df["z"] = df["z"].str.replace("Decimal","")
df["z"] = df["z"].str.replace("]","")
df["z"] = df["z"].str.replace("(","")
df["z"] = df["z"].str.replace(")","")
df["z"] = df["z"].str.replace("'","")
```

```python
for ind in df.index:
    i = df['rotation'][ind]
    x = df['x'][ind]
    y = df['y'][ind]
    z = df['z'][ind]
    
    l = df['l'][ind]
    w = df['w'][ind]
    h = df['h'][ind]
    
    if i == 0:
        a = l
        b = w
        c = h
    elif i == 1:
        a = w
        b = l
        c = h
    elif i == 2:
        a = h
        b = w
        c = l
    elif i == 3:
        a = w
        b = h
        c = l
    elif i == 4:
        a = h
        b = l
        c = w
    elif i == 5:
        a = l
        b = h
        c = w
        
    print(x,y,z,a,b,c)
    
    str_cube = x,y,z,a,b,c
    cuboid_str.append(str_cube)
```
